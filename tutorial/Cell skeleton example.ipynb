The goal of this notebook is to create a cell “Cell skeleton” which has given inputs and outputs.

In NaaVRE, inputs are variables assigned in a cell _above_ and used in the cell. Outputs are variables assigned in the cell and used in a cell _below_.

To create cells, start by opening the component containerizer in the left:




## Set inputs

This cell creates two variables which are consumed in the next one. If you select it and click on “Analyze cell”, you will see that it has two outputs:



```python
# Set inputs

input_1 = "An input variable"
input_2 = "Another input variable"
```

## Cell skeleton

This is our main cell. It consumes the input variables, and creates new output variables. 



```python
# Cell skeleton

# read inputs
print(input_1)
print(input_2)

# ... compute things

# set outputs
output_1 = "An output variable"
output_2 = "Another output variable"
```

## Consume outputs

This cell consumes the output variables. Without it, the above cell would show no output.



```python
# Consume outputs
print(output_1)
print(output_2)
```
