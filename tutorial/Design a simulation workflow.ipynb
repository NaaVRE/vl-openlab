### Create draft components
To create a draft component for your workflow, do the following:
1) Select the notebook cell and open the NaaVRE containerizer:
2) Press `Analyze cell`
3) In `Base image` select `python`
4) Check the `Draft` checkbox
5) Press Create

### Compose a workflow
To use the draft cells to design a workflow, do the following:
1) Open en new launcher:
2) Open the experiment manager:
3) Find the workflow component you have created in the catalog:
4) Drag and drop the cell onto the canvas:
Workflow contain multiple components. Create another draft component before continuing with the following step:
5) Drag and drop the second component onto the canvas:
6) Connect the two blocks:
You have succesfully build a draft workflow. Continue creating draft components and adding them to the workflow to design a workflow that you could use in your own research.
