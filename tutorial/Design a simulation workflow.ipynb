### Create draft components
To create a draft component for your workflow, do the following:
1) Select the notebook cell and open the NaaVRE containerizer:<br/>
<img src="./images/containerizer.png" alt="containerizer.png" width="600"/>
2) Press `Analyze cell`
3) In `Base image` select `python`
4) Check the `Draft` checkbox:<br/>
<img src="./images/draft_checkbox.png" alt="draft_checkbox.png" width="500"/>
5) Press Create

### Compose a workflow
To use the draft cells to design a workflow, do the following:
1) Open a new launcher:<br/>
<img src="./images/new_launcher.png" alt="new_launcher.png" width="600"/> 
2) Open the experiment manager:<br/>
<img src="./images/experiment_manager.png" alt="experiment_manager.png" width="300"/>
3) Find the workflow component you have created in the catalog.
4) Drag and drop the cell onto the canvas.
<br/><br/>
Workflow contain multiple components. Create another draft component before continuing with the following step:
5) Drag and drop the second component onto the canvas.
6) Connect the two blocks.
You have succesfully build a draft workflow.
<br/><br/>
